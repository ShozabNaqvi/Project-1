```python
import os
import pandas as pd
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi
from MCForecastTools import MCSimulation
```

```python
load_dotenv()
```

```python
###Set Alpaca API key and secret
alpaca_api_key = os.getenv("api_key_id")
alpaca_secret_key = os.getenv("secret_api_key")
###Create the Alpaca API object
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
###Set the tickers
tickers = ["TSLA"]
###Set timeframe to '1D' for Alpaca API
timeframe = "1D"
# Set start and end datetimes 
start_date = pd.Timestamp("2015-01-01", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2020-08-27", tz="America/New_York").isoformat()

# Get historical data 
df_ticker = alpaca.get_barset(
    tickers,
    timeframe,
    start=start_date,
    end=end_date
).df

df_ticker.tail(10)
```

```python
# Set number of simulations
num_sims = 1000

# Configure a Monte Carlo simulation to forecast 5 years daily returns
MC_5y = MCSimulation(
    portfolio_data = df_ticker,
    weights = [1],
    num_simulation = num_sims,
    num_trading_days = 252*5
)
MC_5y.portfolio_data.head()
```

```python
MC_5y.calc_cumulative_return()
```

```python
 # Plot probability distribution and confidence intervals
dist_plot_5y = MC_5y.plot_distribution()

# Save the plot for future usage
dist_plot_5y.get_figure().savefig('MC_fiveyear_dist_plot.png',bbox_inches='tight')
```

```python
 # Fetch summary statistics from the Monte Carlo simulation results
tbl_5y = MC_5y.summarize_cumulative_return()

# Print summary statistics
print(tbl_5y)
```

```python
 # Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $10,000 investments in Tesla stocks
ci_lower_5y = round(tbl_5y[8]*10000,2)
ci_upper_5y = round(tbl_5y[9]*10000,2)

# Print results
print(f"There is a 95% chance that an initial investment of $10,000 in the portfolio"
      f" over the next 5 years will end within the range of"
      f" ${ci_lower_5y} and ${ci_upper_5y}")
```
